GROUP BY --- HAVING

GROUP BY works after the WHERE clause filters the rows and before the HAVING clause filters the groups.

- Count Employees by JobTitle
- Average TotalPay by JobTitle
- Employees with the Highest TotalPay in Each Year
- Employees with TotalPay Greater Than 1,000,000 by JobTitle
- Employees with Average BasePay Greater Than 50,000 by JobTitle
- Employee Count in Each Year

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries group by jobtitle'''

sql_query(q)
```

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries 
    group by jobtitle
    where jobtitle != 'Not provided'
    '''

sql_query(q)
```

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries
    where jobtitle != 'Not provided'
    group by jobtitle
    order by 1
    '''

sql_query(q)
```

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries
    where jobtitle != 'Not provided'
    group by jobtitle
    order by 1 desc
    limit 10
    '''

sql_query(q)
```

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries
    where jobtitle != 'Not provided'
    group by jobtitle
    order by 1 desc
    limit 10
    '''

x = sql_query(q)
x.plot(kind='bar', x='JobTitle', y='Employee Count', color='red', figsize=(10,6))
```

```python
#Count Employees by JobTitle
q = '''select count(employeename) 'Employee Count', Jobtitle from 
    salaries
    where jobtitle != 'Not provided'
    group by jobtitle
    order by 1 desc
    limit 10
    '''

x = sql_query(q)
x.plot(kind='barh', x='JobTitle', y='Employee Count', color='green', figsize=(10,6))
```

```python
# Average TotalPay by JobTitle
q = '''select avg(totalpay) 'Average Pay', jobtitle from salaries
    group by 2'''

sql_query(q)
```

```python
# Average TotalPay by JobTitle and overtime pay > 50000 excluded not provided
q = '''select avg(totalpay) 'Average Pay', jobtitle from salaries
    where jobtitle != 'Not provided'
    group by 2
    having overtimepay > 50000
    '''

sql_query(q)
```

### case

Categorizing Employees Based on overtimepay > 100000 'overtime worker' else 'regular worker'

```python
q = '''select employeename, overtimepay,
    case
    when overtimepay > 100000 then "overtime worker"
    else "regular worker"
    end as "work category"
    from salaries'''

sql_query(q)
    
    
```

Categorizing Job Titles Based on totalpay > 100000 'executive level', > 50k and < 99999 'management level' , else 'officer level'

```python
q = '''select jobtitle, totalpay,
    case
    when totalpay > 100000 then 'executive level'
    when totalpay between 50000 and 99999 then 'management level'
    else 'officer level'
    end as 'work category'
    from salaries
    '''

sql_query(q)
```

```python
q = '''select jobtitle, totalpay,
    case
    when totalpay > 100000 then 'executive level'
    when totalpay between 50000 and 99999 then 'management level'
    else 'officer level'
    end as 'work category'
    from salaries
    where jobtitle != 'Not provided'
    
    '''

sql_query(q)
```

```python
q = '''select jobtitle, totalpay,
    case
    when totalpay > 100000 then 'executive level'
    when totalpay between 50000 and 99999 then 'management level'
    else 'officer level'
    end as 'work category'
    from salaries
    where jobtitle != 'Not provided'
    limit 100
    '''

sql_query(q)
```
