Drop rows where EmployeeName and JobTitle are 'Not provided'

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Salaries - salaries.csv')

rec_1 = df[~((df['EmployeeName'] == 'Not provided') & (df['JobTitle'] == 'Not provided'))]

rec_1
```

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

```python
rec_2 = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']] = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']].replace('Not Provided', 0)

rec_2
```

Change BasePay	OvertimePay	OtherPay	Benefits type to float

```python
rec_3 = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']] = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']].astype(float)

rec_3
```

Display all employees name and basepay columns who have a BasePay greater than 200,000.

```python
rec_4 = df[df['BasePay'] > 200000][['EmployeeName', 'BasePay']]

rec_4
```

Find all employees whose JobTitle contains the word “Manager” (case = False)

```python
rec_5 = df[df['JobTitle'].str.contains('Manager', case=False, na=False)]

rec_5
```

Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.

```python
rec_6 = df[['EmployeeName', 'JobTitle', 'TotalPay']]

rec_6
```

Find count of employees who have zero OvertimePay.

```python
rec_7 = df[df['OvertimePay'] == 0]

rec_7
```

Calculate the TotalPay for each JobTitle where totalpay > 300000

```python
rec_8 = df[df['TotalPay'] > 300000].groupby('JobTitle')['TotalPay'].sum()

rec_8
```

Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order

```python
rec_9 = df.groupby('JobTitle')['TotalPayBenefits'].mean().sort_values(ascending=False).head(5)

rec_9
```

Replace all missing Benefits values with 0

```python
df['Benefits'] = df['Benefits'].fillna(0)
df.head()
```

Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe

```python
rec_10 = df.loc[df.groupby('Year')['TotalPay'].idxmax(), ['Year', 'EmployeeName', 'TotalPay']]

rec_10
```

Find the number of employees per JobTitle.

```python
rec_11 = df['JobTitle'].value_counts()

rec_11
```
