```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("TMeURO.csv")
df
```

```python
df.dtypes
```

```python
df.drop(["Latitude","Longtitude"],axis=1,inplace=True)

df
```

```python
df.rename(columns={'City Code': 'city','Order Priority Code':'Order priorty'}, inplace=True)
df
```

```python
df.duplicated()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated()
```

```python
df.isnull().sum()
```

```python
city= df["city"].median()
city
```

```python
orderp= df["Order priorty"].median()
orderp
```

```python
df['city'].replace(to_replace = np.nan, value= city, inplace = True)
df['Order priorty'].replace(to_replace = np.nan, value=orderp, inplace = True)
df
```

```python
df.isnull().sum()
```

```python
city_map = {
    1: "Karachi",
    2: "Lahore",
    3: "Islamabad",
    4: "Faisalabad",
    5: "Quetta",
    6: "Peshawar"
}

# Replace numeric codes with city names
df["city"] = df["city"].map(city_map)
df
```

```python

top5 = df.groupby("product")["total"].sum().sort_values(ascending=False).head(5).reset_index()

sns.barplot(data=top5, x="product", y="total")
plt.title("Top 5 Best Selling Products")
plt.show()
```

Observation:
Electronics and Fashion are our heavy hitters. They bring in the most revenue, while Food is the lowest performer among the top sellers. We need to keep stocking and marketing the top two categories to keep the cash flowing.

```python
counts = df["payment_method"].value_counts()

plt.pie(counts.values, labels=counts.index, autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.show()
```

Observation:
Our customers prefer digital wallets! JazzCash (33.4%) and Easypaisa (28.8%) combine for over 60% of all transactions. Cash on Delivery (COD) is still popular at 20%, but digital is dominant. We should optimize our payment system for JazzCash/Easypaisa and run promotions to encourage even more digital payments.

```python
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Sort months correctly
month_sales = df.groupby("Month")["total"].sum().reset_index()

month_sales["Month"] = pd.Categorical(
    month_sales["Month"],
    categories=month_order,
    ordered=True
)

month_sales = month_sales.sort_values("Month")

# Plot
plt.plot(month_sales["Month"], month_sales["total"], marker="o")
plt.title("Monthwise Sales")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

Observation:
We are growing strong! Sales show a healthy, positive trend over the last 17 months, with our biggest spike happening in March 2024. This consistent year-over-year growth confirms market acceptance and shows our efforts are scaling successfully.

```python
city_sales = df.groupby("city")["total"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=city_sales, y="city", x="total")
plt.title("city-wise Sales (Horizontal Bar Chart)")
plt.xlabel("Total Sales")
plt.ylabel("city")
plt.tight_layout()
plt.show()
```

Observation:
Karachi is our most valuable market. It leads significantly in total sales. Faisalabad is a strong second, but Islamabad lags behind. We should invest heavily in marketing and operations in Karachi to maximize that opportunity, while figuring out how to boost sales in Islamabad.



```python
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Make 'Month' a categorical variable with the proper order
df["Month"] = pd.Categorical(df["Month"], categories=month_order, ordered=True)

# Create pivot table
pivot = df.pivot_table(index="Month", columns="category", values="total",
                       aggfunc="sum", fill_value=0)

# Sort index (months) in calendar order
pivot = pivot.sort_index()

# Plot
pivot.plot(kind="bar", figsize=(10,6))
plt.title("Monthly Sales by Category (Month-wise)")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category")
plt.show()
```

Observation:
Electronics is the engine of the business, every month. It consistently dominates all sales, with Fashion maintaining a solid second place. The growth rate for both Electronics and Fashion is the steepest. Our primary focus must be on ensuring these two core categories are always well-stocked and promoted.

conclusion:

The business is scaling successfully, driven by digital payments and key products in Karachi. Electronics and Fashion are the core revenue drivers, showing the steepest growth in sales, which peaked strongly in March 2024. Karachi is the high-priority market, contributing the largest sales volume. Digital wallets (JazzCash/Easypaisa) account for over 60% of all transactions. These insights require optimizing inventory for the top two categories, focusing marketing budgets on Karachi, and enhancing the digital payment infrastructure to maintain growth momentum.
